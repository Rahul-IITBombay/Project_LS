```python
import gradio as gr
from transformers import pipeline

# Create the sentiment analysis pipeline using XLNet
classifier = pipeline("sentiment-analysis", model="xlnet-base-cased")

# Define the function for Gradio
def analyze_sentiment(tweet):
    sentiment = classifier(tweet)[0]
    label = sentiment['label']
    score = sentiment['score']

    if label == 'LABEL_1':
        sentiment_text = "Positive"
    else:
        sentiment_text = "Negative"

    return f"Sentiment: {sentiment_text} (Score: {score:.2f})"

# Description for the Gradio interface
description = "This is an AI sentiment analyzer that predicts the sentiment of a tweet using XLNet."

# Create the Gradio interface
app = gr.Interface(fn=analyze_sentiment, inputs="text", outputs="text", title="XLNet Sentiment Analyzer", description=description)

# Launch the Gradio interface
app.launch()
```
